```python
# =========================================================
# STEP 3A: Helper Functions for Preprocessing
# =========================================================

def load_rgb_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def load_mask(path):
    mask = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    return mask

def extract_green_channel(img):
    # img shape: H x W x 3 (RGB)
    green = img[:, :, 1]
    return green

def apply_clahe(gray_img):
    clahe = cv2.createCLAHE(
        clipLimit=2.0,
        tileGridSize=(8, 8)
    )
    enhanced = clahe.apply(gray_img)
    return enhanced

def normalize_image(img):
    img = img.astype(np.float32) / 255.0
    return img
```

```python
# =========================================================
# STEP 4: Positive Pixel Ratio Check
# =========================================================

def compute_positive_pixel_ratio(mask_dir):
    total_pixels = 0
    vessel_pixels = 0

    mask_files = sorted(os.listdir(mask_dir))

    for file_name in tqdm(mask_files):
        mask_path = os.path.join(mask_dir, file_name)

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        # Binary vessel mask
        mask = (mask > 0).astype(np.uint8)

        vessel_pixels += np.sum(mask)
        total_pixels += mask.size

    ratio = vessel_pixels / total_pixels
    background_ratio = 1.0 - ratio

    return ratio, background_ratio


vessel_ratio, background_ratio = compute_positive_pixel_ratio(TRAIN_MASK_DIR)

print("\n==============================")
print("Dataset Pixel Distribution")
print("==============================")
print(f"Vessel Pixel Ratio     : {vessel_ratio:.6f}")
print(f"Background Pixel Ratio : {background_ratio:.6f}")
print("==============================")
```

```python
# =========================================================
# STEP 5B: Custom DRIVE Dataset
# =========================================================

class DriveDataset(Dataset):
    def __init__(
        self,
        img_dir,
        mask_dir,
        transform=None
    ):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.transform = transform

        self.image_files = sorted(os.listdir(img_dir))
        self.mask_files = sorted(os.listdir(mask_dir))

        assert len(self.image_files) == len(self.mask_files), \
            "Mismatch between images and masks"

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        # -----------------------------
        # File paths
        # -----------------------------
        img_path = os.path.join(
            self.img_dir,
            self.image_files[idx]
        )

        mask_path = os.path.join(
            self.mask_dir,
            self.mask_files[idx]
        )

        # -----------------------------
        # Load RGB image
        # -----------------------------
        image = cv2.imread(img_path)
        image = cv2.cvtColor(
            image,
            cv2.COLOR_BGR2RGB
        )

        # -----------------------------
        # Green channel extraction
        # -----------------------------
        image = image[:, :, 1]

        # -----------------------------
        # CLAHE enhancement
        # -----------------------------
        clahe = cv2.createCLAHE(
            clipLimit=2.0,
            tileGridSize=(8, 8)
        )
        image = clahe.apply(image)

        # -----------------------------
        # Load mask
        # -----------------------------
        mask = cv2.imread(
            mask_path,
            cv2.IMREAD_GRAYSCALE
        )

        # Binary vessel mask
        mask = (mask > 0).astype(np.float32)

        # -----------------------------
        # Albumentations
        # -----------------------------
        if self.transform:
            augmented = self.transform(
                image=image,
                mask=mask
            )
            image = augmented["image"]
            mask = augmented["mask"]

        # Ensure shape: [1, H, W]
        if len(mask.shape) == 2:
            mask = mask.unsqueeze(0)

        return image, mask
```

```python
# =========================================================
# FIXED STEP 6A: Conv Block for Opacus Compatibility
# =========================================================

class ConvBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()

        self.block = nn.Sequential(
            nn.Conv2d(
                in_channels,
                out_channels,
                kernel_size=3,
                padding=1,
                bias=False
            ),
            nn.GroupNorm(
                num_groups=8,
                num_channels=out_channels
            ),
            nn.ReLU(inplace=False),  # FIXED

            nn.Conv2d(
                out_channels,
                out_channels,
                kernel_size=3,
                padding=1,
                bias=False
            ),
            nn.GroupNorm(
                num_groups=8,
                num_channels=out_channels
            ),
            nn.ReLU(inplace=False)   # FIXED
        )

    def forward(self, x):
        return self.block(x)
```

```python
# =========================================================
# STEP 6B: Full DP-Friendly UNet
# =========================================================

class DPUNet(nn.Module):
    def __init__(self, in_channels=1, out_channels=1):
        super().__init__()

        # Encoder
        self.enc1 = ConvBlock(in_channels, 32)
        self.pool1 = nn.MaxPool2d(2)

        self.enc2 = ConvBlock(32, 64)
        self.pool2 = nn.MaxPool2d(2)

        self.enc3 = ConvBlock(64, 128)
        self.pool3 = nn.MaxPool2d(2)

        self.enc4 = ConvBlock(128, 256)
        self.pool4 = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = ConvBlock(256, 512)

        # Decoder
        self.up4 = nn.ConvTranspose2d(
            512, 256,
            kernel_size=2,
            stride=2
        )
        self.dec4 = ConvBlock(512, 256)

        self.up3 = nn.ConvTranspose2d(
            256, 128,
            kernel_size=2,
            stride=2
        )
        self.dec3 = ConvBlock(256, 128)

        self.up2 = nn.ConvTranspose2d(
            128, 64,
            kernel_size=2,
            stride=2
        )
        self.dec2 = ConvBlock(128, 64)

        self.up1 = nn.ConvTranspose2d(
            64, 32,
            kernel_size=2,
            stride=2
        )
        self.dec1 = ConvBlock(64, 32)

        # Final output
        self.final = nn.Conv2d(
            32,
            out_channels,
            kernel_size=1
        )

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        p1 = self.pool1(e1)

        e2 = self.enc2(p1)
        p2 = self.pool2(e2)

        e3 = self.enc3(p2)
        p3 = self.pool3(e3)

        e4 = self.enc4(p3)
        p4 = self.pool4(e4)

        # Bottleneck
        b = self.bottleneck(p4)

        # Decoder
        d4 = self.up4(b)
        d4 = torch.cat([d4, e4], dim=1)
        d4 = self.dec4(d4)

        d3 = self.up3(d4)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)

        return self.final(d1)
```

```python
# =========================================================
# STEP 7A: Soft Dice Loss
# =========================================================

class SoftDiceLoss(nn.Module):
    def __init__(self, smooth=1.0):
        super().__init__()
        self.smooth = smooth

    def forward(self, logits, targets):
        # IMPORTANT:
        # sigmoid only here (not inside model)
        probs = torch.sigmoid(logits)

        probs = probs.view(-1)
        targets = targets.view(-1)

        intersection = (probs * targets).sum()

        dice = (
            2.0 * intersection + self.smooth
        ) / (
            probs.sum() +
            targets.sum() +
            self.smooth
        )

        return 1.0 - dice
```

```python
# =========================================================
# STEP 7B: Hybrid Loss
# =========================================================

# From your dataset statistics
vessel_ratio = 0.086316
background_ratio = 0.913684

# Strong positive weighting
# helps vessel pixels get enough importance
POS_WEIGHT = background_ratio / vessel_ratio

print("Using pos_weight =", round(POS_WEIGHT, 4))


class HybridLoss(nn.Module):
    def __init__(self):
        super().__init__()

        self.bce = nn.BCEWithLogitsLoss(
            pos_weight=torch.tensor(
                [POS_WEIGHT],
                device=device
            )
        )

        self.dice = SoftDiceLoss()

    def forward(self, logits, targets):
        bce_loss = self.bce(
            logits,
            targets
        )

        dice_loss = self.dice(
            logits,
            targets
        )

        # Weighted hybrid
        total_loss = (
            0.7 * bce_loss +
            0.3 * dice_loss
        )

        return total_loss


criterion = HybridLoss()

print("Hybrid loss ready.")
```

```python
# =========================================================
# STEP 9A: Metric Functions
# =========================================================

def dice_score(pred, target, smooth=1e-6):
    pred = pred.reshape(-1)
    target = target.reshape(-1)

    intersection = (pred * target).sum()

    return (
        2.0 * intersection + smooth
    ) / (
        pred.sum() + target.sum() + smooth
    )


def iou_score(pred, target, smooth=1e-6):
    pred = pred.reshape(-1)
    target = target.reshape(-1)

    intersection = (pred * target).sum()
    union = pred.sum() + target.sum() - intersection

    return (
        intersection + smooth
    ) / (
        union + smooth
    )


def binary_metrics(pred, target):
    pred = pred.reshape(-1)
    target = target.reshape(-1)

    tn, fp, fn, tp = confusion_matrix(
        target,
        pred,
        labels=[0, 1]
    ).ravel()

    sensitivity = tp / (tp + fn + 1e-6)
    specificity = tn / (tn + fp + 1e-6)
    accuracy = (tp + tn) / (
        tp + tn + fp + fn + 1e-6
    )

    return sensitivity, specificity, accuracy
```

```python
# =========================================================
# STEP 9B: Validation Threshold Search
# =========================================================

def validate_model(model, loader):
    model.eval()

    thresholds = np.arange(0.20, 0.81, 0.05)

    best_threshold = 0.5
    best_dice = 0.0

    best_metrics = {
        "dice": 0.0,
        "iou": 0.0,
        "auc": 0.0,
        "sensitivity": 0.0,
        "specificity": 0.0,
        "accuracy": 0.0
    }

    with torch.no_grad():
        all_probs = []
        all_targets = []

        for images, masks in loader:
            images = images.to(device)
            masks = masks.to(device)

            logits = model(images)
            probs = torch.sigmoid(logits)

            all_probs.append(
                probs.cpu().numpy()
            )
            all_targets.append(
                masks.cpu().numpy()
            )

        all_probs = np.concatenate(all_probs, axis=0)
        all_targets = np.concatenate(all_targets, axis=0)

        # AUC uses probabilities
        auc = roc_auc_score(
            all_targets.reshape(-1),
            all_probs.reshape(-1)
        )

        for threshold in thresholds:
            preds = (
                all_probs > threshold
            ).astype(np.uint8)

            dice = dice_score(
                preds,
                all_targets
            )

            if dice > best_dice:
                best_dice = dice
                best_threshold = threshold

                iou = iou_score(
                    preds,
                    all_targets
                )

                sens, spec, acc = binary_metrics(
                    preds,
                    all_targets.astype(np.uint8)
                )

                best_metrics = {
                    "dice": dice,
                    "iou": iou,
                    "auc": auc,
                    "sensitivity": sens,
                    "specificity": spec,
                    "accuracy": acc
                }

    return best_threshold, best_metrics
```

```python
# =========================================================
# STEP 10A: Training Configuration
# =========================================================

EPOCHS = 200
TARGET_EPSILON = 50.0
DELTA = 1e-5

SAVE_PATH = "best_drive_dp_model.pth"

best_val_dice = 0.0
best_threshold = 0.5

patience = 20
early_stop_counter = 0

print("Training config ready.")
```

```python
# =========================================================
# STEP 10B: Full Training Loop
# + Epsilon Checkpoint Saving
# =========================================================

import os
import time
from tqdm import tqdm

print("=" * 70)
print("DP Retinal Vessel Segmentation Training Started")
print("=" * 70)

# =========================================================
# Epsilon checkpoint setup
# =========================================================

TARGET_EPSILONS = [1, 5, 10, 20, 30, 40, 50]

SAVE_EPS_DIR = "epsilon_checkpoints"
os.makedirs(SAVE_EPS_DIR, exist_ok=True)

saved_epsilons = set()

print("Epsilon checkpoint tracking enabled.")

# =========================================================
# Training Loop
# =========================================================

for epoch in range(EPOCHS):
    model.train()

    train_loss = 0.0
    start_time = time.time()

    # ----------------------------------
    # Training
    # ----------------------------------
    for images, masks in tqdm(train_loader):
        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad()

        logits = model(images)

        loss = criterion(
            logits,
            masks
        )

        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    avg_train_loss = train_loss / len(train_loader)

    # ----------------------------------
    # Validation
    # ----------------------------------
    val_threshold, metrics = validate_model(
        model,
        val_loader
    )

    val_dice = metrics["dice"]

    # ----------------------------------
    # Privacy epsilon
    # ----------------------------------
    epsilon = privacy_engine.get_epsilon(
        delta=DELTA
    )

    # ----------------------------------
    # Save checkpoint at target epsilon
    # ----------------------------------
    for target_eps in TARGET_EPSILONS:

        if (
            epsilon >= target_eps
            and target_eps not in saved_epsilons
        ):

            eps_save_path = os.path.join(
                SAVE_EPS_DIR,
                f"model_eps_{target_eps}.pth"
            )

            torch.save(
                model.state_dict(),
                eps_save_path
            )

            saved_epsilons.add(
                target_eps
            )

            print(
                f"★ Checkpoint saved at ε={target_eps}"
            )

    epoch_time = (
        time.time() - start_time
    ) / 60.0

    # ----------------------------------
    # Print epoch results
    # ----------------------------------
    print(
        f"Epoch [{epoch+1:03d}/{EPOCHS}] | "
        f"Loss={avg_train_loss:.4f} | "
        f"Dice={metrics['dice']:.4f} | "
        f"IoU={metrics['iou']:.4f} | "
        f"AUC={metrics['auc']:.4f} | "
        f"Sens={metrics['sensitivity']:.4f} | "
        f"Spec={metrics['specificity']:.4f} | "
        f"Acc={metrics['accuracy']:.4f} | "
        f"Thr={val_threshold:.2f} | "
        f"ε={epsilon:.2f} | "
        f"t={epoch_time:.2f} min"
    )

    # ----------------------------------
    # Save best overall model
    # ----------------------------------
    if val_dice > best_val_dice:
        best_val_dice = val_dice
        best_threshold = val_threshold
        early_stop_counter = 0

        torch.save(
            model.state_dict(),
            SAVE_PATH
        )

        print(
            f"★ Best model saved "
            f"(Dice={best_val_dice:.4f})"
        )

    else:
        early_stop_counter += 1

    # ----------------------------------
    # Early stopping
    # ----------------------------------
    if early_stop_counter >= patience:
        print("\nEarly stopping triggered.")
        break

    # ----------------------------------
    # Privacy budget control
    # ----------------------------------
    if epsilon >= TARGET_EPSILON:
        print(
            f"\nStopped: epsilon exceeded "
            f"target ({TARGET_EPSILON})"
        )
        break

print("\nTraining completed.")
print(f"Best Validation Dice : {best_val_dice:.4f}")
print(f"Best Threshold       : {best_threshold:.2f}")
print(f"Saved Model          : {SAVE_PATH}")
print(f"Epsilon Checkpoints  : {SAVE_EPS_DIR}")
```

```python
# =========================================================
# FINAL TEST EVALUATION
# Dice + IoU + Accuracy + Sensitivity + Specificity
# + AUC + MCC + ROC Curve
# For BASELINE or DP model
# =========================================================

from sklearn.metrics import (
    roc_auc_score,
    matthews_corrcoef,
    roc_curve
)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch

# ---------------------------------------------------------
# Load Best Model
# Change path if needed
# ---------------------------------------------------------

MODEL_PATH = "best_drive_baseline_model.pth"
# For DP model use:
# MODEL_PATH = "best_drive_dp_model.pth"

model = DPUNet(
    in_channels=1,
    out_channels=1
).to(device)

model.load_state_dict(
    torch.load(
        MODEL_PATH,
        map_location=device
    )
)

model.eval()

print("Model loaded from:", MODEL_PATH)


# ---------------------------------------------------------
# Evaluation Function
# ---------------------------------------------------------

def final_test_evaluation(
    model,
    loader,
    threshold=0.70
):
    model.eval()

    all_probs = []
    all_preds = []
    all_targets = []

    with torch.no_grad():
        for images, masks in loader:
            images = images.to(device)
            masks = masks.to(device)

            logits = model(images)

            probs = torch.sigmoid(
                logits
            )

            preds = (
                probs > threshold
            ).float()

            all_probs.extend(
                probs.cpu().numpy().flatten()
            )

            all_preds.extend(
                preds.cpu().numpy().flatten()
            )

            all_targets.extend(
                masks.cpu().numpy().flatten()
            )

    all_probs = np.array(all_probs)
    all_preds = np.array(all_preds)
    all_targets = np.array(all_targets)

    # -------------------------------------------------
    # Metrics
    # -------------------------------------------------

    intersection = (
        all_preds * all_targets
    ).sum()

    dice = (
        (2.0 * intersection + 1e-8)
        /
        (
            all_preds.sum()
            + all_targets.sum()
            + 1e-8
        )
    )

    iou = (
        (intersection + 1e-8)
        /
        (
            all_preds.sum()
            + all_targets.sum()
            - intersection
            + 1e-8
        )
    )

    tp = (
        (all_preds == 1)
        &
        (all_targets == 1)
    ).sum()

    tn = (
        (all_preds == 0)
        &
        (all_targets == 0)
    ).sum()

    fp = (
        (all_preds == 1)
        &
        (all_targets == 0)
    ).sum()

    fn = (
        (all_preds == 0)
        &
        (all_targets == 1)
    ).sum()

    accuracy = (
        (tp + tn)
        /
        (tp + tn + fp + fn + 1e-8)
    )

    sensitivity = (
        tp
        /
        (tp + fn + 1e-8)
    )

    specificity = (
        tn
        /
        (tn + fp + 1e-8)
    )

    auc = roc_auc_score(
        all_targets,
        all_probs
    )

    mcc = matthews_corrcoef(
        all_targets.astype(np.int64),
        all_preds.astype(np.int64)
    )

    return {
        "Dice Score": dice,
        "IoU Score": iou,
        "Accuracy": accuracy,
        "Sensitivity": sensitivity,
        "Specificity": specificity,
        "AUC": auc,
        "MCC": mcc
    }, all_targets, all_probs


# ---------------------------------------------------------
# Run Evaluation
# ---------------------------------------------------------

metrics, y_true, y_prob = final_test_evaluation(
    model,
    test_loader,
    threshold=best_threshold
)

results_df = pd.DataFrame(
    list(metrics.items()),
    columns=["Metric", "Value"]
)

print("\n" + "=" * 60)
print("FINAL TEST RESULTS")
print("=" * 60)
print(
    results_df.to_string(
        index=False
    )
)
print("=" * 60)


# ---------------------------------------------------------
# ROC Curve
# ---------------------------------------------------------

fpr, tpr, _ = roc_curve(
    y_true,
    y_prob
)

plt.figure(figsize=(8, 6))

plt.plot(
    fpr,
    tpr,
    linewidth=2,
    label=f"ROC Curve (AUC = {metrics['AUC']:.4f})"
)

plt.plot(
    [0, 1],
    [0, 1],
    linestyle="--",
    linewidth=1
)

plt.title(
    "ROC Curve - Retinal Vessel Segmentation"
)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# =========================================================
# FIX FOR OPACUS CHECKPOINT LOADING
# (_module. prefix issue)
# BASELINE vs DP-SGD MCC vs Epsilon
# =========================================================

from sklearn.metrics import matthews_corrcoef
import matplotlib.pyplot as plt
import numpy as np
import os
import torch

TARGET_EPSILONS = [1, 5, 10, 20, 30, 40, 50]

DP_SAVE_DIR = "epsilon_checkpoints"
BASELINE_MODEL_PATH = "best_drive_baseline_model.pth"

print("=" * 70)
print("BASELINE vs DP-SGD MCC COMPARISON")
print("=" * 70)


# =========================================================
# Helper: Remove Opacus "_module." prefix
# =========================================================

def clean_opacus_state_dict(state_dict):
    new_state_dict = {}

    for key, value in state_dict.items():
        if key.startswith("_module."):
            new_key = key.replace("_module.", "")
        else:
            new_key = key

        new_state_dict[new_key] = value

    return new_state_dict


# =========================================================
# MCC Calculation Function
# =========================================================

def calculate_mcc(model, loader, threshold):
    model.eval()

    all_preds = []
    all_targets = []

    with torch.no_grad():
        for images, masks in loader:
            images = images.to(device)
            masks = masks.to(device)

            logits = model(images)
            probs = torch.sigmoid(logits)

            preds = (probs > threshold).float()

            all_preds.extend(
                preds.cpu().numpy().flatten()
            )

            all_targets.extend(
                masks.cpu().numpy().flatten()
            )

    all_preds = np.array(all_preds).astype(np.int64)
    all_targets = np.array(all_targets).astype(np.int64)

    mcc = matthews_corrcoef(
        all_targets,
        all_preds
    )

    return mcc


# =========================================================
# STEP 1: BASELINE MCC
# =========================================================

print("\nCalculating BASELINE MCC...")

baseline_model = DPUNet(
    in_channels=1,
    out_channels=1
).to(device)

baseline_model.load_state_dict(
    torch.load(
        BASELINE_MODEL_PATH,
        map_location=device
    )
)

baseline_mcc = calculate_mcc(
    baseline_model,
    test_loader,
    threshold=best_threshold
)

print(
    f"Baseline MCC = {baseline_mcc:.4f}"
)


# =========================================================
# STEP 2: DP MCC at Each Epsilon
# =========================================================

print("\nCalculating DP-SGD MCC at Different ε values...")

epsilon_values = []
mcc_values = []

for eps in TARGET_EPSILONS:

    checkpoint_path = os.path.join(
        DP_SAVE_DIR,
        f"model_eps_{eps}.pth"
    )

    if not os.path.exists(checkpoint_path):
        print(f"Checkpoint missing for ε={eps}")
        continue

    dp_model = DPUNet(
        in_channels=1,
        out_channels=1
    ).to(device)

    raw_state_dict = torch.load(
        checkpoint_path,
        map_location=device
    )

    cleaned_state_dict = clean_opacus_state_dict(
        raw_state_dict
    )

    dp_model.load_state_dict(
        cleaned_state_dict
    )

    dp_mcc = calculate_mcc(
        dp_model,
        test_loader,
        threshold=best_threshold
    )

    epsilon_values.append(eps)
    mcc_values.append(dp_mcc)

    print(
        f"ε={eps} | MCC={dp_mcc:.4f}"
    )


# =========================================================
# STEP 3: Plot MCC vs Epsilon
# =========================================================

plt.figure(figsize=(10, 6))

# DP curve
plt.plot(
    epsilon_values,
    mcc_values,
    marker="o",
    linewidth=2,
    label="DP-SGD Model"
)

# Point labels
for x, y in zip(epsilon_values, mcc_values):
    plt.text(
        x,
        y + 0.01,
        f"{y:.4f}",
        ha="center",
        fontsize=9
    )

# Baseline horizontal line
plt.axhline(
    y=baseline_mcc,
    linestyle="--",
    linewidth=2,
    label=f"Baseline (MCC={baseline_mcc:.4f})"
)

plt.title(
    "Baseline vs DP-SGD: MCC vs Privacy Budget"
)

plt.xlabel(
    "Privacy Budget (ε)"
)

plt.ylabel(
    "Matthews Correlation Coefficient (MCC)"
)

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
